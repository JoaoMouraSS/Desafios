```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
data_train = pd.read_csv('train.csv')
data_test = pd.read_csv('test.csv')
```

```python
data_train.head()
```

```python
# Modify in train data

data_train['Sex_Binario'] = data_train['Sex'].map({'male':0, 'female':1})
data_train.drop('Sex', axis = 1, inplace=True)
```

```python
data_train.fillna(-1, inplace=True)
```

```python
data_train.head()
```

```python
# Modify in test data

data_test['Sex_Binario'] = data_test['Sex'].map({'male':0, 'female':1})
data_test.drop('Sex', axis = 1, inplace=True)
data_test.fillna(-1, inplace=True)
```

```python
# Selecting the features for training

x_train = data_train[['Sex_Binario','Age','Pclass', 'SibSp', 'Parch', 'Fare']]
y_train = data_train['Survived']
```

```python
# separeting data

from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x_train,y_train, random_state = 111, test_size = 0.20,stratify=y_train)
```

```python
modelo1 = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=111)
modelo1.fit(train_x, train_y)
```

```python
predict_test = modelo1.predict(test_x)
```

```python
# accuracy_score - train and validation

from sklearn.metrics import accuracy_score

print('Taxa de acerto é:', accuracy_score(test_y, predict_test))
```

```python
x_test = data_test[['Sex_Binario','Age','Pclass', 'SibSp', 'Parch', 'Fare']]
x_test
```

```python
y = modelo1.predict(x_test)
```

```python
# Saving the model on computer for submision

results = pd.Series(y, index=data_test['PassengerId'], name='Survived')

results.to_csv('second_model.csv', header = True)
```

```python
# Testing another model

from sklearn.svm import SVC

modelo2 = SVC()

modelo2.fit(train_x, train_y)
```

```python
# Accuracy of this new model

predict_test2 = modelo1.predict(test_x)

print('The accuracy of this model is:', accuracy_score(test_y, predict_test2))
```

```python
# Another model

from sklearn.ensemble import AdaBoostClassifier

modelo3 = AdaBoostClassifier()
modelo3.fit(train_x, train_y)

# Accuracy of this model

predict_test3 = modelo3.predict(test_x)

print('The accuracy os this model is:', accuracy_score(test_y, predict_test3))
```
